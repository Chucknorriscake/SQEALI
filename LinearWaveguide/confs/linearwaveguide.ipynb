#### Coupling distance over waveguide distances

- a) calculate with default values for the minimum finding
- b) change fit values if default is not enough
- c) plot relation

```python
import numpy as np
coupling_lengths= []
x_axis = []
for file in os.listdir():
    if ("dist" in file) and ("ez" not in file):
        parts = file.split("_")[1].split(".")[0:2]
        combined = ".".join(parts)
        combined = (float(combined))
        linear = LinearWaveguides(file)
        if linear.ez_loaded == False:
            linear.run_simulation(save_ez=True)
        coupling_lengths.append(linear.calc_coupling_length())
        x_axis.append(combined)
```

```python
linear = LinearWaveguides("dist_1.3.json")
linear.plot_coupling_length()
```

```python
plt.scatter(x_axis, coupling_lengths)
plt.xlabel("waveguide distance")
plt.ylabel("coupling length")
```
